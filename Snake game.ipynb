```python
from tkinter import Tk
from tkinter import *
import tkinter
import random
```

```python
ROW = 25
COLUMN = 25
TILE_SIZE = 25
WINDOW_WIDTH = TILE_SIZE * ROW
WINDOW_HEIGHT = TILE_SIZE * COLUMN

# Create window
window = tkinter.Tk()
window.title("Snake")
window.resizable(False, False)

canvas = tkinter.Canvas(window, bg="black", height=WINDOW_HEIGHT, width=WINDOW_WIDTH, borderwidth=0, highlightthickness=0)
canvas.pack()
window.update()

# Center the window on the screen
def center_window():
    window_width = window.winfo_width()
    window_height = window.winfo_height()
    screen_width = window.winfo_screenwidth()
    screen_height = window.winfo_screenheight()
    x = int((screen_width / 2) - (window_width / 2))
    y = int((screen_height / 2) - (window_height / 2))
    window.geometry(f"{window_width}x{window_height}+{x}+{y}")

center_window()

# Tile class
class Tile:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Game initialization
snake = Tile(5 * TILE_SIZE, 5 * TILE_SIZE)
food = Tile(10 * TILE_SIZE, 10 * TILE_SIZE)
snake_body = []
velocityX, velocityY = 0, 0
game_over = False
score = 0

# Change snake direction
def change_direction(event):
    global velocityX, velocityY

    if event.keysym == "Right" and velocityX != -1:
        velocityX = 1
        velocityY = 0
    elif event.keysym == "Left" and velocityX != 1:
        velocityX = -1
        velocityY = 0
    elif event.keysym == "Up" and velocityY != 1:
        velocityX = 0
        velocityY = -1
    elif event.keysym == "Down" and velocityY != -1:
        velocityX = 0
        velocityY = 1

# Move the snake
def move():
    global snake, food, game_over, snake_body, score

    if game_over:
        return

    # Check for collisions with walls
    if snake.x < 0 or snake.x >= WINDOW_WIDTH or snake.y < 0 or snake.y >= WINDOW_HEIGHT:
        game_over = True
        return

    # Check for collisions with itself
    for tile in snake_body:
        if snake.x == tile.x and snake.y == tile.y:
            game_over = True
            return

    # Check if snake has eaten the food
    if snake.x == food.x and snake.y == food.y:
        snake_body.append(Tile(food.x, food.y))
        food.x = random.randint(0, COLUMN - 1) * TILE_SIZE
        food.y = random.randint(0, ROW - 1) * TILE_SIZE
        score += 1

    # Update snake body
    if snake_body:
        snake_body = [Tile(snake.x, snake.y)] + snake_body[:-1]

    # Move the snake head
    snake.x += velocityX * TILE_SIZE
    snake.y += velocityY * TILE_SIZE

# Draw game elements
def draw():
    global game_over, score

    move()
    canvas.delete("all")

    # Draw the snake
    canvas.create_rectangle(snake.x, snake.y, snake.x + TILE_SIZE, snake.y + TILE_SIZE, fill="lime green")
    
    # Draw the food
    canvas.create_rectangle(food.x, food.y, food.x + TILE_SIZE, food.y + TILE_SIZE, fill="red")
    
    # Draw the snake body
    for tile in snake_body:
        canvas.create_rectangle(tile.x, tile.y, tile.x + TILE_SIZE, tile.y + TILE_SIZE, fill="lime green")
    
    # Game Over text
    if game_over:
        canvas.create_text(WINDOW_WIDTH / 2, WINDOW_HEIGHT / 2, font="Arial 20", text=f"Game Over, Score: {score}", fill="white")
    else:
        canvas.create_text(30, 20, font="Arial 20", text=f"Score: {score}", fill="white")
    
    window.after(100, draw)

# Start the game
draw()
window.bind("<KeyRelease>", change_direction)
window.mainloop()
```
